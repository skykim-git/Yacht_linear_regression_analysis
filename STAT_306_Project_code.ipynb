```r
yacht <- read.csv("yacht_hydro.csv")
name <- c("LC","PC","LD","BD","LB","FR","RR")
colnames(yacht) <- name
head(yacht,3)
library(ggplot2)
library(GGally)
library(leaps)
```

```r
model1 <- lm(RR~. ,data = yacht)
summary(model1)
plot(model1$fitted.values,model1$residuals,xlab = "fitted value",ylab = "residual",main=("Residual vs Fitted"))
abline(a = 0,b = 0)
```

### Model 4 5 6

```r
s <- regsubsets(log(RR) ~LC+PC+LD+BD+LB+FR+I(FR^2)+I(FR^3)+I(FR^4), data = yacht, method="exhaustive",nvmax = 9) 
ss <- summary(s)
ss$which
round(ss$adjr2,4)
round(ss$cp,4)
```

```r
p <- c(1,2,3,4,5,6,7,8,9)
plot(p,ss$cp,xlab = "number of parameters",ylab = "Cp",main = "Mallow's Cp Statistic vs Parameter Number")
abline(a = 0,b= 1,col = "red")
plot(p,ss$adjr2,xlab = "number of parameters",ylab = "adjusted R-square",main = "Adjusted R-square vs Parameter Number")
```

```r
## error mod 4
train <- 1:as.integer(dim(yacht)[1]/2)## number of datasets?(integer?)
reg1 <- lm(log(RR)~LC+PC+BD+FR+I(FR^2)+I(FR^3)+I(FR^4), data=yacht[train,])## regress on the first half
error1 <- sum((log(yacht$RR)[-train] - predict(reg1, yacht[-train,]))^2)## check on the other half
reg2 <- lm(log(RR)~LC+PC+BD+FR+I(FR^2)+I(FR^3)+I(FR^4), data=yacht[-train,]) ## regress on the second half
error2 <- sum((log(yacht$RR)[train] - predict(reg2, yacht[train,]))^2) ## check on the other half
error <- (error1 + error2)/dim(yacht)[1]
error
```
